```python
import numpy as np 
import pandas as pd 
from sklearn.linear_model import LinearRegression 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("your_file.csv")
print(df.head())
```

```python
X = df[['Experience', 'Education', 'Age']]
y = df['Salary']
```

```python
model = LinearRegression()
model.fit(X, y)
```

```python
print("Intercept:", model.intercept_)
print("Coefficients:", model.coef_)
```

```python
input_data = [[6, 16, 30]]
predicted_salary = model.predict(input_data)

print("Predicted Salary:", predicted_salary[0])
```

```python
plt.scatter(df['Experience'], y, color='blue', label='Actual')
plt.scatter(df['Experience'], model.predict(X), color='red', label='Predicted')
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("MLR without Train-Test Split")
plt.legend()
plt.show()
```

```python
# Fix Education & Age at their mean values
edu_mean = df['Education'].mean()
age_mean = df['Age'].mean()

# Vary Experience smoothly
exp_range = np.linspace(df['Experience'].min(), df['Experience'].max(), 100)

X_fixed = pd.DataFrame({
    'Experience': exp_range,
    'Education': [edu_mean]*100,
    'Age': [age_mean]*100
})

y_pred_fixed = model.predict(X_fixed)

# Plot
plt.scatter(df['Experience'], y, color='blue', label='Actual')
plt.plot(exp_range, y_pred_fixed, color='red', linewidth=2, label='Predicted (MLR)')
plt.xlabel("Experience")
plt.ylabel("Salary")
plt.title("CORRECT MLR Visualization (Others Fixed)")
plt.legend()
plt.show()
```
